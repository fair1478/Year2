```python
# Data Modelling using Regression Techniques
X = df.drop('Price', axis = 1)
y = df['Price']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=106)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
rSq = model.score(X_train,y_train)
rSq
```

```python
out = model.predict(X_test)
```

```python
p = pd.DataFrame(X_test)
p["Price"] = y_test
p["predicted"] = out
p["Error"] = ((out/y_test) -1)*100
p
```
